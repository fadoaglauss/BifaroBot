## Critérios de exclusão de robôs

Obedecendo as boas praticas, sempre antes de iniciar a coleta de uma pagina ,verificamos se a mesma permite ser coletada. Caso não , ignoramos e buscamos outra url. Caso a pagina permita , iremos buscar seu conteudo e fazer a busca por novas urls. Repetindo este processo. Para verificamos se a pagina permite ser coletada, usamos o RobotFileParser da biblioteca urllib. Esta classe tem um metodo que nos permite verificar por meio do robots.txt se a pagina permite ou não sua coleta.
